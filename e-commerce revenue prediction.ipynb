```python
y = df1.iloc[:,18]
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_tr,x_te,y_tr,y_te = train_test_split(x,y,test_size = 0.3, random_state = 42)
```

```python
print(x_tr.shape)
print(x_te.shape)
print(y_tr.shape)
print(y_te.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
df1.isna().sum()
```

```python
lr = LogisticRegression()
lr_model = lr.fit(x_tr,y_tr)
lr_pred = lr_model.predict(x_te)
#accuracy_score(y_te,lr_pred)
```

```python
accuracy_score(y_te,lr_pred)
```

```python
print(classification_report(y_te,lr_pred))
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_tr_sc = sc.fit_transform(x_tr)
x_te_sc = sc.fit_transform(x_te)
```

```python
lr_sc = LogisticRegression()
lr_model_sc = lr_sc.fit(x_tr_sc,y_tr)
lr_pred_sc = lr_model_sc.predict(x_te_sc)
accuracy_score(y_te,lr_pred_sc)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_sc = RandomForestClassifier()
rf_model_sc = rf_sc.fit(x_tr_sc,y_tr)
rf_pred_sc = rf_model_sc.predict(x_te_sc)
accuracy_score(y_te,rf_pred_sc)
```
