```python
X=df2.loc[:,ind].values
```

```python
y=df2['Churn']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=50,stratify = y)
```

```python
print(X_train.shape,y_train.shape)
```

```python
print(X_test.shape,y_test.shape)
```

```python
log_reg = LogisticRegression(class_weight = 'balanced').fit(X_train,y_train)
```

```python
pred = log_reg.predict(X_test)
```

```python
print(pred)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,pred,target_names = ['non-churn','churn']))
```

```python
confusion_matrix(y_test,pred)
```

```python
from sklearn.metrics import roc_auc_score,roc_curve,auc
```

```python
false_positive_rates,true_positive_rates,thresholds = roc_curve(y_test,pred)
```

```python
roc_auc = auc(false_positive_rates,true_positive_rates)
```

```python
roc_auc
```

```python
plt.plot(false_positive_rates,true_positive_rates)
plt.xlabel(false_positive_rates)
plt.ylabel(true_positive_rates)
plt.show()
```
