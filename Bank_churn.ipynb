```python
df1 = df[['RowNumber', 'CreditScore', 'Geography',
       'Gender', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard',
       'IsActiveMember', 'EstimatedSalary', 'Exited']]
```

```python
df1.head()
```

```python
df1.dtypes
```

```python
df1.columns
```

```python
df2 = df1[['RowNumber', 'CreditScore', 'Geography', 'Gender', 'Age', 'Tenure',
       'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember',
       'EstimatedSalary']]
```

```python
x = np.asarray(df2)
```

```python
y = np.asarray(df1['Exited'])
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_tr,x_te,y_tr,y_te = train_test_split(x,y,test_size = 0.3,random_state = 100,stratify = y)
```

```python
print(x_tr.shape,y_tr.shape)
```

```python
print(x_te.shape,y_te.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train_sc = sc.fit_transform(x_tr)
x_test_sc = sc.transform(x_te)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
dt_model = DecisionTreeClassifier(criterion= 'entropy', random_state=0)
dt_model.fit(x_train_sc, y_tr)
y_pred_dt_sc = dt_model.predict(x_test_sc)
accuracy_score(y_te, y_pred_dt_sc)
```

```python
cr = classification_report(y_te, y_pred_dt_sc)
 
print("Classification report >>> \n", cr)
```

```python
#Logistic regression
lr_model2 = LogisticRegression(random_state = 0)
lr_model2.fit(x_train_sc, y_tr)
y_pred_lr_sc = lr_model2.predict(x_test_sc)
accuracy_score(y_te, y_pred_lr_sc)
```

```python
cr = classification_report(y_te, y_pred_lr_sc)
 
print("Classification report >>> \n", cr)
```

```python
#Random forest classifier
rf_model2 = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0)
rf_model2.fit(x_train_sc, y_tr)
y_pred_rf_sc = rf_model2.predict(x_test_sc)
accuracy_score(y_te, y_pred_rf_sc)
```

```python
cr = classification_report(y_te, y_pred_rf_sc)
 
print("Classification report >>> \n", cr)
```

```python
#KNN classifier
knn_model2 = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2,)
knn_model2.fit(x_train_sc, y_tr)
y_pred_knn_sc = knn_model2.predict(x_test_sc)
accuracy_score(y_te, y_pred_knn_sc)
```

```python
cr = classification_report(y_te, y_pred_knn_sc)
 
print("Classification report >>> \n", cr)
```

```python
#SVM classifier
svc_model2 = SVC()
svc_model2.fit(x_train_sc, y_tr)
y_pred_svc_sc = svc_model2.predict(x_test_sc)
 
accuracy_score(y_te, y_pred_svc_sc)
```

```python
cr = classification_report(y_te, y_pred_svc_sc)
 
print("Classification report >>> \n", cr)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ada_mod = AdaBoostClassifier(n_estimators = 50,learning_rate=0.1)

model = ada_mod.fit(x_train_sc, y_tr)

y_pred_ada_sc = model.predict(x_test_sc)

accuracy_score(y_te, y_pred_ada_sc)
```

```python
cr = classification_report(y_te, y_pred_ada_sc)
 
print("Classification report >>> \n", cr)
```
