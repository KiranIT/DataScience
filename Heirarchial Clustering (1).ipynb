```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("clustering_dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
x=df.loc[:,:].values
```

```python
#import libraries to plot dendrogram
import scipy.cluster.hierarchy as sch
```

```python
#plot dedrogram to visualise the possbile number of clusters
dendogram=sch.dendrogram(sch.linkage(x,method='ward'))
```

```python
plt.figure(figsize=(10,7))
plt.title("Dendrogram")
dendogram=sch.dendrogram(sch.linkage(x,method='ward'))
plt.axhline(y=1700,color='r',linestyle='--')
plt.show()
```

```python
## Training entire data using Heirarchical Clustering
from sklearn.cluster import AgglomerativeClustering 
```

```python
agg=AgglomerativeClustering(n_clusters=3)
```

```python
cluster=agg.fit(x)
```

```python
cluster.labels_
```

```python
#visualize the clusters  with labels
plt.scatter(x[:,0],x[:,1], c=cluster.labels_, cmap='rainbow')  
plt.show()
```
